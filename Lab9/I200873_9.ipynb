### Necessary Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading Data

```python
data = pd.read_csv("price-prediction.csv")
print(data.head(5))
```

### Visualize Datapoints in Data

```python
plt.scatter(data['PLOTS'],data['PRICE'])
```

### Some Base Functions

```python
def plot_data(m, b, data):
    data.plot(kind = 'scatter', x = 'PLOTS', y = 'PRICE')
    minval = data['PLOTS'].min()
    maxval = data['PLOTS'].max()
    plt.plot([x for x in range(minval, maxval)], [ (m*x+b) for x in range(minval, maxval)], color='red')
    plt.xlabel('PLOTS')
    plt.ylabel('PRICE')
    plt.show()
```

```python
def loss_function(m, c, points):
    total_err = float(0)
    x = points['PLOTS']
    y = points['PRICE']    
    for i,j in enumerate(x):
        y1 = m*j+c
        total_err = (y[i]-y1)**2
    total_err /= data['PLOTS'].count()
    return total_err
```

```python
# Formula of slope: -1/N * x * (y - (m * x + b))
# Formula of y_intercept: -1/N * (y - (m * x + b))

def gradient_descent(current_m, current_b, points, lr):
    m = 0
    b = 0
    countx = points['PLOTS'].count()
    for i in range(countx):
        x = points.iloc[i].PLOTS
        y = points.iloc[i].PRICE
        m += -1/countx * x * (y - (current_m * x + current_b))
        b += -1/countx * (y - (current_m * x + current_b))
    
    newm = current_m-(lr*m)
    newb = current_b-(lr*b)
    
    return newm,newb
```

### Running Script

```python
# Initialize the values of m and b to 0
m = 0
b = 0
# Set the learning rate and number of epochs for gradient descent
lr = 0.000001
epochs = 500
# Perform gradient descent for the specified number of epochs
for i in range(epochs):
    # Update m and b using gradient descent
    m,b = gradient_descent(m,b,data,lr)
    # Plot the current data
    if(i%50)==0:
        plot_data(m,b,data)
        input("Press key to continue")
# Print the final values of m and b
print("M: "+str(m))
print("B: "+str(b))
```

### Predicting New Data

```python
# Read a datapoint as 'x' from the user
plotsize = float(input("Enter Plot Size:   "))
# Does a prediction by calculating mx+b
predict = m*plotsize+b
# Print the prediction
print(f'Price for {plotsize} size of plot should be around $:{predict}')
```
