```python
goal = [[1,2,3],[4,5,6],[7,8,0]]
#puzzle = [[0,1,3],[4,2,5],[7,8,6]]
#puzzle = [[5,6,7],[4,0,8],[3,2,1]] # Worst
puzzle = [[1,3,2],[7,8,4],[6,0,5]] # Medium
#puzzle = [[1,3,4],[8,6,2],[7,0,5]] # Easy
```

```python
def checkvisited(node):
    if node and (node in visited):
        return True
    return False
def printPuzzle(board):
    for row in board:
        print("-------------")
        print("|",row[0],"|",row[1],"|" , row[2] , "|")
    print("-------------")
    print()
def isSolved(node):
    return node == goal
def getZero(node):
    row = 0
    col = 0
    for i in range(3):
        for j in range(3):
            if node[i][j] == 0:
                row,col = i,j
                return row,col
```

```python
def heuristicFunction(node):
    
    distance = 0
    for i in range(len(goal)):
        for j in range(len(goal)):
            if (node[i][j] != goal[i][j]):
                row = node[i][j] //3
                col = node[i][j] %3
                distance += abs(row-i)+abs(col-i)
    return distance
```

```python
visited = list()
nextstates = PriorityQueue()
```

```python
from queue import PriorityQueue

# Define the goal state
goal_state = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

# Define the heuristic function
def heuristic(state):
    count = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != goal_state[i][j]:
                count += 1
    return count

# Define the A* algorithm function
def A_star(initial_state):
    frontier = PriorityQueue()
    frontier.put((0, initial_state))
    explored = set()
    while not frontier.empty():
        current_state = frontier.get()[1]
        print(current_state)
        if current_state == goal_state:
            print(current_state)
            return True
        explored.add(tuple(map(tuple, current_state)))
        for move in get_possible_moves(current_state):
            new_state = make_move(current_state, move)
            if tuple(map(tuple, new_state)) not in explored:
                priority = heuristic(new_state) + 1
                frontier.put((priority, new_state))
    return False

# Define the helper function to get possible moves
def get_possible_moves(state):
    moves = []
    row, col = get_blank_position(state)
    if row > 0:
        moves.append((row - 1, col))
    if row < 2:
        moves.append((row + 1, col))
    if col > 0:
        moves.append((row, col - 1))
    if col < 2:
        moves.append((row, col + 1))
    return moves

# Define the helper function to make a move
def make_move(state, move):
    new_state = [row[:] for row in state]
    row, col = move
    new_state[row][col], new_state[get_blank_position(state)[0]][get_blank_position(state)[1]] = new_state[get_blank_position(state)[0]][get_blank_position(state)[1]], new_state[row][col]
    return new_state

# Define the helper function to get the position of the blank tile
def get_blank_position(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return (i, j)

# Test the function
initial_state = [[1, 2, 3], [0, 4, 6], [7, 5, 8]]
print(A_star(initial_state))
```
