```python
import copy
from queue import PriorityQueue
```

```python
class NQueens:
    def __init__(self, n):
        self.n = n
        self.board = []
        for i in range(n):   
            col = []
            for j in range(n): 
                col.append("*")   
            self.board.append(col)
        
    def place_queen(self, row , c ):
        col = int(0)
        added = int(0)
        if (self.board[row][c] != "x" and self.board[row][c] != "Q" ):
            self.board[row][c] = "Q"
            col = c
            added = 1
        if added == 0:
            return 0
        
        #print("Queen Placed in Col {}".format(col))
        # Cross Row
        for i in range(self.n):
            if (self.board[row][i] == "*"):
                self.board[row][i] = "x"
            i+=1
        # Cross Col
        for i in range(self.n):
            if (self.board[i][col] == "*"):
                self.board[i][col] = "x"
                
                
        # Cross Main Diagonal
        checkcol = col
        i = row
        while (i<self.n) and (checkcol<self.n):
            if (self.board[i][checkcol] == "*"):
                self.board[i][checkcol] = "x"
            checkcol+=1
            i+=1
            
        
        # Cross Diagonal 2
        checkcol = col
        i = row
        while (i>=0) and (checkcol>=0):
            if (self.board[i][checkcol] == "*"):
                self.board[i][checkcol] = "x"
            checkcol-=1
            i-=1
        # Cross Anti Diagonal
        checkcol = col
        i = row
        while (i>=0) and (checkcol<self.n):
            if (self.board[i][checkcol] == "*"):
                self.board[i][checkcol] = "x"
            checkcol+=1
            i-=1
        # Cross Anti Diagonal 2
        checkcol = col
        i = row
        while (i<self.n) and (checkcol>=0):
            if (self.board[i][checkcol] == "*"):
                self.board[i][checkcol] = "x"
            checkcol-=1
            i+=1
        return 1
    def getQueens(self):
        queencount =   int(0)
        for i in range(self.n):
            for j in range (self.n):
                if self.board[i][j] == "Q":
                    queencount+=1
        return queencount      
 
    def display_board(self):
        for i in range(self.n):
            for j in range(self.n):
                print(self.board[i][j], end=' ')
            print()
```

```python
population = []
#population = PriorityQueue()
size = int(input("Enter Board Size: "))
q = NQueens(size)
#population.put((size,copy.deepcopy(q)))
population.append(copy.deepcopy(q))

iteration = int(0)
solutionset = []
while len(population)>0:
#while not population.empty():
    q = copy.deepcopy(population.pop())
    iteration = q.getQueens()
    for i in range(size):
        r = copy.deepcopy(q)
        res = r.place_queen(iteration,i)
        
        if r.getQueens() == size:
            if (r not in solutionset):
                solutionset.append(copy.deepcopy(r))
        elif res == 1:
            population.append(copy.deepcopy(r))
            # Priority Queue and Heuristic
            
            #h = size-r.getQueens()
            #population.put((h,r))
            
print('-'*30)
print("Total Solution Found: {}".format(len(solutionset)))
print('-'*30)
print("Solution Set")            
for i in solutionset:
    i.display_board()
    print('-'*30)
```
