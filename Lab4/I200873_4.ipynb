```python
import copy
from queue import PriorityQueue
```

```python
goal = [[1,2,3],[4,5,6],[7,8,0]]
#puzzle = [[0,1,3],[4,2,5],[7,8,6]]
#puzzle = [[5,6,7],[4,0,8],[3,2,1]] # Worst
puzzle = [[1,3,2],[7,8,4],[6,0,5]] # Medium
#puzzle = [[1,3,4],[8,6,2],[7,0,5]] # Easy
```

```python
def printPuzzle(board):
    for row in board:
        print("-------------")
        print("|",row[0],"|",row[1],"|" , row[2] , "|")
    print("-------------")
    print()
def isSolved(node):
    return node == goal
def getZero(node):
    row = 0
    col = 0
    for i in range(3):
        for j in range(3):
            if node[i][j] == 0:
                row,col = i,j
                return row,col
```

```python
printPuzzle(goal)
printPuzzle(puzzle)

print(isSolved(puzzle))
print(getZero(puzzle))
```

```python
visited = list()
nextstates = list()
```

```python
def checkvisited(node):
    if (node in visited):
        return True
    return False
```

```python
def lefttopcorner(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r+1][c]
    p1[r+1][c] = num

    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r][c+1]
    p2[r][c+1] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    
def righttopcorner(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r][c-1]
    p1[r][c-1] = num

    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r+1][c]
    p2[r+1][c] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    
def topcenter(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r][c-1]
    p1[r][c-1] = num

    p3 = copy.deepcopy(node)
    num = p3[r][c]
    p3[r][c] = p3[r][c+1]
    p3[r][c+1] = num
    
    
    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r+1][c]
    p2[r+1][c] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    try:
        if checkvisited(p3) == False:
            nextstates.append(p3)
    except:
        print("Already in List")

def leftbottomcorner(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r-1][c]
    p1[r-1][c] = num

    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r][c+1]
    p2[r][c+1] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    
def rightbottomcorner(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r][c-1]
    p1[r][c-1] = num

    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r-1][c]
    p2[r-1][c] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    
def bottomcenter(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r][c-1]
    p1[r][c-1] = num

    p3 = copy.deepcopy(node)
    num = p3[r][c]
    p3[r][c] = p3[r][c+1]
    p3[r][c+1] = num
    
    
    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r-1][c]
    p2[r-1][c] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    try:
        if checkvisited(p3) == False:
            nextstates.append(p3)
    except:
        print("Already in List")

def leftmidcorner(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r-1][c]
    p1[r-1][c] = num
    
    p3 = copy.deepcopy(node)
    num = p3[r][c]
    p3[r][c] = p3[r+1][c]
    p3[r+1][c] = num

    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r][c+1]
    p2[r][c+1] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    try:
        if checkvisited(p3) == False:
            nextstates.append(p3)
    except:
        print("Already in List")
    
def rightmidcorner(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r][c-1]
    p1[r][c-1] = num

    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p2[r][c] = p2[r-1][c]
    p2[r-1][c] = num
    
    p3 = copy.deepcopy(node)
    num = p3[r][c]
    p3[r][c] = p3[r+1][c]
    p3[r+1][c] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    try:
        if checkvisited(p3) == False:
            nextstates.append(p3)
    except:
        print("Already in List")
    
def midcenter(node):
    r,c = getZero(node)
    p1 = copy.deepcopy(node)
    num = p1[r][c]
    p1[r][c] = p1[r][c-1]
    p1[r][c-1] = num

    p3 = copy.deepcopy(node)
    num = p3[r][c]
    p3[r][c] = p3[r][c+1]
    p3[r][c+1] = num
    
    
    p2 = copy.deepcopy(node)
    num = p2[r][c]
    p1[r][c] = p1[r-1][c]
    p1[r-1][c] = num
    
    p4 = copy.deepcopy(node)
    num = p4[r][c]
    p4[r][c] = p4[r+1][c]
    p4[r+1][c] = num
    
    try:
        if checkvisited(p1) == False:
            nextstates.append(p1)
    except:
        print("Already in List")
    try:
        if checkvisited(p2) == False:
            nextstates.append(p2)
    except:
        print("Already in List")
    try:
        if checkvisited(p3) == False:
            nextstates.append(p3)
    except:
        print("Already in List")
    try:
        if checkvisited(p4) == False:
            nextstates.append(p4)
    except:
        print("Already in List")
        
```

```python
nextstates.append(puzzle)
```

```python
printPuzzle(puzzle)
while len(nextstates)>0:
    if len(nextstates) == 0:
        print("Couldn't Find Solution")
        break
    node = nextstates.pop()
    
    if(node in visited):
        continue
        
    visited.append(copy.deepcopy(node))
    r ,c = getZero(node)
    if (node == goal):
        print("Goal Found")
        break

    if (r == 0):
        if (c == 0):
            lefttopcorner(node)
        if (c == 1):
            topcenter(node)
        if (c == 2):
            righttopcorner(node)
    if (r == 1):
        if (c == 0):
            leftmidcorner(node)
        if (c == 1):
            midcenter(node)
        if (c == 2):
            rightmidcorner(node)
    if (r == 2):
        if (c == 0):
            leftbottomcorner(node)
        if (c == 1):
            bottomcenter(node)
        if (c == 2):
            rightbottomcorner(node)
    
    
printPuzzle(node)
```
